```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data_Caso_Propuesto.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
categoricas = df.select_dtypes(include=['object'])
numericas = df.select_dtypes(include=['int64', 'float64'])

print("Columnas Categóricas:")
print(categoricas.columns)

print("Columnas Numéricas:")
print(numericas.columns)
```

```python
df.isnull().sum().sum()
```

```python
duplicados = df.duplicated()
print(duplicados.any())
```

```python
columnas_con_faltantes = df.columns[df.isnull().any()]
print(columnas_con_faltantes)
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
plt.hist(df['Precio'], bins=10)
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencia de Precio')
plt.show()
```

```python
# Crear una matriz de correlación
correlation_matrix = df.corr(numeric_only=True)

# Generar el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
```

```python
plt.scatter(df['Precio'], df['Area Terreno'])
plt.show()
```

```python
plt.scatter(df['Precio'], df['Area Construida'])
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(data=df, x='Area Terreno')
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.countplot(data=df, x='Area Construida')
plt.show()
```

```python
# Supongamos que quieres eliminar los valores de '0' en la columna 'columna_limpiar'
df = df[df['Area Terreno'] != 0]

# Generar la gráfica con los datos limpios
plt.plot(df['Area Terreno'])

# Agregar etiquetas y título al gráfico
plt.xlabel('Índice')
plt.ylabel('Valores de la columna')
plt.title('Gráfico de la columna con datos limpios')

# Mostrar el gráfico
plt.show()
```
